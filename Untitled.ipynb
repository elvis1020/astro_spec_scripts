```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib nbagg
```

```python
global mt
global ag
global cE
global cMm
global pcurve
global ptxt
global fidu_met
```

```python
def CMbuild(filein):
    data = np.genfromtxt(filein, unpack=True)
    age = data[1]
    isochrones = {}
    lastage = 0
    for n, i in enumerate(age):
        if i != lastage:
            isochrones[i] = [[], []]
        B = data[24][n]
        V = data[25][n]
        if V < 25:
            isochrones[i][0].append(B)
            isochrones[i][1].append(V)
        lastage = i
    return isochrones
```

```python
def HRbuild(filein):
    data = np.genfromtxt(filein, unpack=True)
    age = data[1]
    isochrones = {}
    lastage = 0
    for n, i in enumerate(age):
        if i != lastage:
            isochrones[i] = [[], []]
        L = data[4][n]
        Teff = data[5][n]
        if data[9][n] > 0.4:
            isochrones[i][0].append(L)
            isochrones[i][1].append(Teff)
        lastage = i
    return isochrones
```

```python
def plotiso(ax, met, age, E, Mm, Vlim):
    global ptxt
    global pcurve
    try:
        pcurve.remove()
        ax.texts.remove(ptxt)
        plt.draw()
    except:
        pass
    Av = 3.1*E
    B = np.array(metset[met][age][0]) + Mm + Av
    V = np.array(metset[met][age][1]) + Mm + Av
    BV = B - V + E
    pcurve = ax.plot(BV, V, 'r-')[0]
    ax.set_ylim([Vlim, min(V)-2])
    distance = 10**((Mm + 5) / 5)
    ptxt = ax.text(0.01, 0.01, met+'; %6.3fGyr; E(B-V)=%4.3f; (M-m)0=%3.1f; d=%7.1fpc'%(age/1E9, E, Mm, distance), transform=ax.transAxes, color='green')
```

```python
def plothr(ax, met, age, style, Vlim):
    L = np.array(metset[met][age][0])
    T = np.array(metset[met][age][1])
    pcurve = ax.plot(T, L, style)[0]
```

```python
def fiducial(ax, met, E, Mm, Vlim):
    global fidu_met
    try:
        for i in fidu_met:
            i.remove()
    except:
        pass
    Av = 3.1*E
    fidu_met = []
    for a in ages:
        B = np.array(metset[met][a][0]) + Mm + Av
        V = np.array(metset[met][a][1]) + Mm + Av
        BV = B - V + E
        fidu_met.append(ax.plot(BV, V, 'k-', alpha=0.05)[0])
    ax.set_ylim([Vlim, min(V)-2])
```

```python
def stepmet(direct, ax):
    global mt
    mets = sorted(metset.keys())
    idx = mets.index(mt)
    if direct == 'up':
        mt = mets[idx + 1]
    if direct == 'down':
        mt = mets[idx - 1]
    pcurve = plotiso(axcm, mt, ag, cE, cMm, lim)
    print mt
```

```python
def stepage(direct, ax):
    global ag
    idx = ages.index(ag)
    if direct == 'up':
        ag = ages[idx + 1]
    if direct == 'down':
        ag = ages[idx - 1]
    pcurve = plotiso(axcm, mt, ag, cE, cMm, lim)
    print ag
```

```python
def setpars(excess, dist, ax):
    global cE
    global cMm
    cE = excess
    cMm = dist
    pcurve = plotiso(axcm, mt, ag, cE, cMm, lim)
```
